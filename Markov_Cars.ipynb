```python
# equations

# 1) using p_k, q_k, r_k to represent p_k+1, q_k+1, r_k+1
# p_k+1 = 0.6p_k + 0.1q_k + 0.1r_k
# q_k+1 = 0.1p_k + 0.8q_k + 0.2r_k
# r_k+1 = 0.3p_k + 0.1q_k + 0.7r_k

# 2)
#  |0.6 0.1 0.1||p_k|   |p_k+1|
#  |0.1 0.8 0.2||q_k| = |q_k+1|
#  |0.3 0.1 0.7||r_k|   |r_k+1|

# 3)
from numpy import array, matmul
M = array([[0.6, 0.1, 0.1], [0.1,0.8, 0.2], [0.3, 0.1, 0.7]])
x_k = array([[200],[200],[200]])
for i in range(30):
    print("After {} week, the cars will be: ".format(i+1))
    x_k = matmul(M, x_k)
    print("Number of Cars = \n", x_k)


# Summary: From the result, we could see the sequence converges. The number of cars will be 120, 270, 210 in the P, Q, R.
# The starting distribution doesn't matter after sufficient number of weeks.

# 4) 

import numpy as np
from numpy.linalg import eig

M = array([[0.6, 0.1, 0.1], [0.1,0.8, 0.2], [0.3, 0.1, 0.7]])
w,v = eig(M)
print("First pair: ", w[0],v[:,0])
print("Second pair: ", w[1],v[:,1])
print("Third pair: ", w[2],v[:,2])


# using the eigvector of the eigvalue equals 1 and x0 to get the x_infnity since the eigvalue of the Marcov matrix is always 1  
x_0 = array([[200],[200],[200]])
B = v[:,0].T
C = 200*3/sum(B)
x_infnity = C * B
print(x_infnity)
```
