```python
from numpy import shape, array
# equations
# -0.2x1 + 0.1x2 + 0x3 + 0x4 + 0x5 + 400 = y1
# 0.1x1 - 0.4x2 + 0x3 + 0.1x4 + 0x5 - 100 = y2
# 0.1x1 + 0.2x2 - 0.5x3 + 0.1x4 + 0x5 - 100 = y3
# 0x1 + 0.1x2 + 0.2x3 - 0.4x4 + 0x5 = y4
# 0x1 + 0x2 + 0.3x3 + 0x4 - 0.5x5 = y5
```

```python

A = array([[-0.2,0.1,0,0,0], [0.1,-0.4,0,0.1,0], [0.1,0.2,-0.5,0.1,0], [0,0.1,0.2,-0.4,0], [0,0,0.3,0,-0.5]])
b = array([[400],[-100],[-100],[0],[0]])
x = array([[400],[400],[400],[400],[400]])

# 1) Fill in the Matrix A and vector b
print("A (row, col):", shape(A))
print("Matrix A is below: ")
for line in A:
    print('  '.join(map(str, line)))

print("b (row, col):", shape(b))
print("Matrix b is below: ")
for line in b:
    print('  '.join(map(str, line)))

print("x (row, col):", shape(x))
print("Matrix x is below: ")
for line in x:
    print('  '.join(map(str, line)))
```

```python
# 2)
# a) change in population for one month
print("Ax = \n", A@x)
print("y = \n", A@x+b)

# b) update the population after first month
print("delta_y = \n", A@x+b+x)
```

```python
# 3) Loop to find population after 12 months
A = array([[-0.2,0.1,0,0,0], [0.1,-0.4,0,0.1,0], [0.1,0.2,-0.5,0.1,0], [0,0.1,0.2,-0.4,0], [0,0,0.3,0,-0.5]])
b = array([[400],[-100],[-100],[0],[0]])
x = array([[400],[400],[400],[400],[400]])
for i in range(12):
    print("After {} month, the population will be: ".format(i+1))
    x = A@x+b+x
    print("population = \n", x)



A = array([[-0.2,0.1,0,0,0], [0.1,-0.4,0,0.1,0], [0.1,0.2,-0.5,0.1,0], [0,0.1,0.2,-0.4,0], [0,0,0.3,0,-0.5]])
b = array([[400],[-100],[-100],[0],[0]])
x = array([[400],[400],[400],[400],[400]])
for i in range(12):
  print("After {} month, the population will be: ".format(i+1))
  x = A@x+b+x
  print("population = \n", x)

```
