```python
#bisection routine

def bisection(f,x1,x2,tol):
    #a)
    if(f(x1)==0 or f(x2)==0):
        print("Error: Initial guesses bracketa root of f(x).")
        return None
    #b)
    if(abs(f(x1))<tol):
        return x1  
    if(abs(f(x2))<tol):
        return x2
    #c)
    x1_n = x1
    x2_n = x2
    best_est = 0;
    
    for n in range(1,101):
        m_n = (x1_n + x2_n)/2
        # print("m_n")
        # print(m_n)
        f_m_n = f(m_n)
        # print("====")
        # print(f_m_n)
        if f(x1_n)*f_m_n < 0:
            x1_n = x1_n
            x2_n = m_n
        elif f(x2_n)*f_m_n < 0:
            x1_n = m_n
            x2_n = x2_n
        elif abs(f_m_n) < tol:
            print("Found exact solution.")
            return m_n
        else:
            print("Bisection method fails.")
            return None
    #d)
    if f(x1)*f(x2) >= 0:
        print("Bisection method fails. x1 and x2 do not span the root.")
        return None
    
    #e)
    if(f(x1)==0):
        return x1  
    if(f(x2)==0):
        return x2
    return (x1_n + x2_n)/2


#1) use bisection function to solve the tank question
D=10
C=0.5
V=45
f = lambda x: pow(D,3)*C*(math.pi/12)*(3*pow(x/D, 2)-2*pow(x/D,3))-V
bisection(f,0,10,0.001)


# 1)Semi-ellipsoidal Tank  

V = sympy.symbols('V')
D=10
C=0.5
h0=2
equation = sympy.Eq(V, pow(D,3)*C*(math.pi/12)*(3*pow(h0/D, 2)-2*pow(h0/D,3)))

V_arbitary=sympy.solve(equation)  
print(V_arbitary) 

V_difference = 45 - V_arbitary[0] 
print(V_difference)


for h in range(h0*1000,10000,2):
    h = h/1000
    V_d = sympy.symbols('V_d')
    equation = sympy.Eq(V_d, pow(D,3)*C*(math.pi/12)*(3*pow(h/D, 2)-2*pow(h/D,3)))
    res = sympy.solve(equation) 
    diff = 45.0-res[0]
    
if(diff < 0.00005):
    print("Answer:")
    print(h)
    break



    
```
